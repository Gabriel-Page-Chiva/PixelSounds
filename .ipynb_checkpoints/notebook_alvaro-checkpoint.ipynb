```python
import funciones_alvaro as fa
import cv2
import os
import numpy as np
import warnings
import scipy.fftpack as fft
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d, median_filter
from scipy import signal
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    import vozyaudio as vz
```

```python
spec = fa.espectrograma_por_bloques(fa.extract_y("images/comaCalvo.png"), block_size=16)

specdb = fa.espectrograma_por_bloques(fa.extract_y("images/comaCalvo.png"), block_size=8,return_db=True)
plt.figure(figsize=(10, 4))
plt.imshow(specdb, aspect='auto', origin='lower', cmap='magma')
plt.colorbar(label='Amplitud (dB)')
plt.xlabel('Bloque (simulado tiempo)')
plt.ylabel('Frecuencia')
plt.title('Espectrograma tipo STFT generado desde imagen')
plt.tight_layout()
plt.show()
```

```python
flatness = fa.flatness(spec)

plt.plot(flatness[0])
plt.title("Flatness espectral por bloque de imagen")
plt.xlabel("Bloque (tiempo simulado)")
plt.ylabel("Flatness")
plt.grid()
plt.show()
```
